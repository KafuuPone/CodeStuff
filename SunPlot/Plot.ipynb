```python
%matplotlib widget
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd
```

```python
output = pd.read_csv("output.csv", header = None)
output.head()
```

```python
x = list(output[0])
y = list(output[1])
z = list(output[2])
```

```python
# Create Figure

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.autoscale(enable = False)
ax.set_xbound(lower=-1, upper=1)
ax.set_ybound(lower=-1, upper=1)
ax.set_zbound(lower=-1, upper=1)
plt.xlabel('x')
plt.ylabel('y')

# Create Plot
ax.scatter3D(x, y, z, color='orange', marker = '.')
```
