```python
import numpy as np
import matplotlib.pyplot as plt

def solar_analemma(delta, alpha, L):
    n = 365
    theta = np.zeros(n)
    x = np.zeros(n)
    y = np.zeros(n)
    for i in range(n):
        theta[i] = (2 * np.pi / 365) * (i - 1)
        x[i] = alpha[i] * np.cos(theta[i])
        y[i] = delta * np.cos(L) * np.cos(theta[i]) - delta * np.sin(L) * np.sin(theta[i])
    return x, y

delta = 23.5 * np.pi / 180
L = np.deg2rad(40.7)
alpha = np.zeros(365)
for i in range(365):
    alpha[i] = (2 * np.pi / 365) * (i - 80)
x, y = solar_analemma(delta, alpha, L)
plt.plot(x, y)
plt.xlabel('Equation of Time (minutes)')
plt.ylabel('Declination (degrees)')
plt.title('Solar Analemma')
plt.grid(True)
plt.show()
```
